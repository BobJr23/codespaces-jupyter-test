```python
import torch, numpy as np, cv2, time
image = cv2.imread(r"/workspaces/codespaces-jupyter-test/data/fn.jpg")
model = torch.hub.load("ultralytics/yolov5", "custom", path = r"/workspaces/codespaces-jupyter-test/data/best.pt")
model.conf = 0.4

for x in range(10):
    t = time.time()
    results = model(image)
    print(time.time() - t)
```

```python
%matplotlib inline

import matplotlib.pyplot as plt

for *box, conf, cls in results.xyxy[0]:
    x1y1 = [int(x.item()) for x in box[:2]]
    x2y2 = [int(x.item()) for x in box[2:]]
    image = cv2.rectangle(image, x1y1, x2y2, (244, 113, 115), 2)
    print(x1y1, x2y2)
print('done')
```
